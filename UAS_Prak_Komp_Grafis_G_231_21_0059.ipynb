```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Data for Purchase Price, Current Price, and Gain/Loss attributes
purchase_price = [90.00, 32.00, 25.00, 50.00, 18.00, 225.00, 22.00, 18.00, 65.00, 800.00, 95.00, 17.00, 45.00]
current_price = [144.13, 65.57, 82.57, 44.56, 17.69, 833.36, 36.07, 33.24, 56.48, 897.64, 86.26, 141.64, 14.61]
gain_loss = [14,269.14, 13,048.57, 12,302.93, 11,094.20, 8,824.82, 82,502.64, 7,176.94, 7,445.54, 10,391.40, 43,984.36, 8,539.74, 141,498.36, 7,287.94]

# Create subplots for each attribute
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Histogram for Purchase Price
axs[0].hist(purchase_price, bins=10, color='blue')
axs[0].set_title('Distribution of Purchase Price')
axs[0].set_xlabel('Price')
axs[0].set_ylabel('Frequency')

# Histogram for Current Price
axs[1].hist(current_price, bins=10, color='green')
axs[1].set_title('Distribution of Current Price')
axs[1].set_xlabel('Price')
axs[1].set_ylabel('Frequency')

# Histogram for Gain/Loss
axs[2].hist(gain_loss, bins=10, color='orange')
axs[2].set_title('Distribution of Gain/Loss')
axs[2].set_xlabel('Amount')
axs[2].set_ylabel('Frequency')

# Adjust spacing between subplots
plt.tight_layout()

# Display the histograms
plt.show()
```
